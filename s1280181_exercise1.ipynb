Reading csv file.

```python
import pandas as pd
import plotly.express as px
df = pd.read_csv("./drive/MyDrive/Ex1/station_Info.csv")
```

modifying dataframe to show map(The "Location" column ).

```python
mid = df['Location']
newDataFrame = pd.DataFrame(columns = ['x','y'])
for td in mid:
  td = td.replace('Point(','')
  td = td.replace(')','')
  td = td.split()
  td = list(map(float,td))
  newDataFrame.loc[len(newDataFrame)] = td

```

execute function to show map.

```python
fg = px.scatter_mapbox(
    df,
    lat = newDataFrame.y,
    lon = newDataFrame.x,
    hover_name = 'rowNumber',
    hover_data=['stationId', 'Location', 'Address'],
    opacity = 1.0,
    zoom=4,
    text = 'Address'
)
```

```python
fg.update_layout(mapbox_style='open-street-map')
fg.update_layout(margin={"r": 0, "t": 0, "l": 0, "b": 0})
fg.update_layout(title_text="Population of each cities")
fg.show()
```
